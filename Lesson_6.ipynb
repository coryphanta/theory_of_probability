**Задача 1**
Даны значения величины заработной платы заемщиков банка (salary) и значения их поведенческого кредитного скоринга (scoring):

salary = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]
scoring = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]

1. Найдите ковариацию этих двух величин: сначала без использования специальных функций, а затем с помощью функции numpy.cov. Полученные значения должны быть равны.
2. Найдите коэффициент корреляции Пирсона: с помощью ковариации и средних квадратических отклонений двух признаков, а затем с помощью специальных функций из библиотек numpy и pandas.



1) $cov_{X,Y} = M(X,Y) - M(X)M(Y)$
<br>
$cov_{X,Y} = 81141.7 - 101.4 \cdot 709.9 = 9157.8399$

```python
import numpy as np

salary = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110] 
scoring = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]

cov_np = np.cov(salary, scoring, ddof=0)[0, 1]
print(cov_np)
```

2) $$r_{XY} = \frac{cov_{XY}}{\sigma_{X}\sigma_{Y}}$$


$\sigma = \sqrt{ \frac{ \sum \limits_{i=1}^{n}(X_{i} - \bar{X})_{i}^2}{n}} $ - Среднеквадратическое отклонение (для знаменателя)
<br>
$\sigma_{x} = 59.1155$
<br>
$\sigma_{y} = 174.5534$
<br>
$r_{XY} = \frac{9157.8399}{59.1155 \cdot 174.5534} = 0.8875$

```python
r1 = cov_np / (np.std(salary, ddof=0) * np.std(scoring, ddof=0))
print(r1)
```

```python
r2 = np.corrcoef(salary, scoring)[0, 1]
print(r2)
```

**Задача 2**

Измерены значения IQ выборки студентов, обучающихся в местных технических вузах:

131, 125, 115, 122, 131, 115, 107, 99, 125, 111

Известно, что в генеральной совокупности IQ распределен нормально. Найдите доверительный интервал для математического ожидания с надежностью 0.95.



$p = 0.95$
<br>
$\mu = 118.1$
<br>
$\sigma = \sqrt{\frac{(131-118.1)^2 + (125 - 118.1)^2 + (115 - 118.1)^2 + (122 - 118.1)^2 + (131 - 118.1)^2 + (115 - 118.1)^2 + (107 - 118.1)^2 + (99 - 118.1)^2 + (125 - 118.1)^2 + (111 - 118.1)^2}{10}} = \sqrt{\frac{166.41 + 47.61 + 9.61 + 15.21 + 166.41 + 9.61 + 123.21 + 364,81 + 47.61 + 50.41}{10}} = 10.05$
<br>
Вычислим уровень значимости
<br>
$\alpha = 1 -p = 0.05$
<br>
Используя формулу $T_{1,2} = \overline{X} \pm t_{\alpha/2} \cdot \frac{s_0}{\sqrt{n}}$, где $t_{\alpha/2} = 2.262$ по табличке распределения Стьюдента, ищем интервал
<br>
$T_{1} = 118.1 - 2.262 \cdot \frac{10.05}{\sqrt{10}} = 110.811$
<br>
$T_{2} = 118.1 + 2.262 \cdot \frac{10.05}{\sqrt{10}} = 125.289$
