### Задача 1

О непрерывной равномерно распределенной случайной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Попробуем найти левую границу через формулу дисперсии:
$D(X) = \frac{(b - a)^2}{12}$

$\frac{(b - 0,5)^2}{12} = 0,2$
<br>
$(b - 0,5)^2 = 12 \cdot 0,2$
<br>
$b - 0,5 = \sqrt{2,4}$
<br>
$b = 2,05$ 
Левая сторона найдена
<br>
Ищем среднее значение
<br>
$M(X) = \frac{a + b}{2}$ - формула
<br>
$M(X) = \frac{0.5 + 2.05}{2} = 1,275$

### Задача 2

Рост взрослого населения города X имеет нормальное распределение, причем, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
1. больше 182 см?
2. больше 190 см?
3. от 166 см до 190 см?
4. от 166 см до 182 см?
5. от 158 см до 190 см?
6. не выше 150 см или не ниже 190 см?
7. не выше 150 см или не ниже 198 см?
8. ниже 166 см?

Задачу можно решить двумя способами: без использования сторонних библиотек (numpy, scipy, pandas и пр.), а затем проверить себя с помощью встроенных функций

M(X) = 174 = a
<br>
$\sigma(X) = 8$
<br>
1) $P(X > 182) = 1 - \left(\left[\Phi\left(\frac{182 - 174}{8}\right)\right]\right) = 1 - \left( \left[\Phi(1)\right] \right) = 1 - 0.8413 = 0.1587$
<br>
2) $P(X > 190) = 1 - \left(\frac{1}{2}\left[\Phi\left(\frac{190 - 174}{8}\right)\right]\right) = 1 - \left( \left[\Phi(2)\right] \right) = 1 - 0.9772 = 0.0228$
<br>
3) $P(166< X <190) = \left(\left[\Phi\left(\frac{190 - 174}{8}\right) - \Phi \left(\frac{166  - 174}{8}\right)\right]\right) = \left( \left[\Phi(2) - \Phi(-1)\right] \right) = 0.9772 - 0.1587 = 0.8186$

8) $P(X < 166) = \left(\left[\Phi\left(\frac{166 - 174}{8}\right)\right]\right) = \left( \left[\Phi(-1)\right] \right) = 0.1587$
