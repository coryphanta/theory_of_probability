**Задача 2**

Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:

202, 203, 199, 197, 195, 201, 200, 204, 194, 190

Известно, что их веса распределены нормально.
1. Верно ли утверждение продавца, если учитывать, что уровень значимости равен 1%?
2. Найдите P-значение для данного теста.



$\bar{X} = 198,5$
<br>
$n = 10$
<br>
Ищем исправленное кв. отклонение
$\sigma = \sqrt{\frac{(x_i - \bar{x})^2}{n-1}} = \sqrt{\frac{178.5}{9}} = 4.453$
<br>
Предположим, что нулевая гипотеза равна:
<br>
$H_o: a = 200$ и противоположная гепотеза $H_1: a \neq 200$
<br>
Ишем T набл.
$T_{набл} = \frac{\bar{X} - a}{\sigma} \cdot \sqrt{n} = \frac{198, 5 - 200}{4,453}\cdot \sqrt{10} = -1,075$
Ишем по табличке критерий Стьюдента = $\alpha = 0,05;$ $n-1 = 9 $ 
<br>
$T_{кр} = 1,833$
<br>
Смотрим доверительный интервал:
<br>
$\left(\bar{X} - T_{kp} \cdot \frac{\sigma}{\sqrt{n}}; \bar{X} + T_{kp} \cdot \frac{\sigma}{\sqrt{n}}\right) = \left(198,5 - 1,833 \cdot\frac{4,453}{\sqrt{10}};198,5 + 1,833 \cdot\frac{4,453}{\sqrt{10}}\right) = (195,92; 201,08)$
<br>
Гипотезе $H_{0}$ можно верить, 200 входит в доверительный интервал.

$(a - 2 \cdot \sigma,  a + 2 \cdot \sigma)$
<br>
$(17 - 2 * 0.2, 17 + 2 * 0,2) = (16.6;17,4)$
<br>
Считаем T(X)
